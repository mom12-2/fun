```python
print("Welcome to the Mad Libs Game! 🎉")
print("Please provide the following words:\n")
# Get user inputs
adjective1 = input("Enter an adjective: ")
noun1 = input("Enter a noun: ")
verb1 = input("Enter a verb (past tense): ")
adjective2 = input("Enter another adjective: ")
noun2 = input("Enter another noun: ")
verb2 = input("Enter a verb: ")

# Create the story
story = f"""
One day, a {adjective1} {noun1} {verb1} through the park.
It was a very {adjective2} day, and everyone was carrying a {noun2}.
Suddenly, the {noun1} decided to {verb2} loudly!

What a strange day!
"""

# Display the story
print("\nHere is your Mad Libs story:")
print(story)
```

```python
print("Welcome to the Restaurant Tip Calculator 🍽️")

# Ask user for inputs
bill = float(input("Enter the total bill amount: $"))
tip_percent = float(input("Enter the tip percentage (e.g., 10, 15, 20): "))

# Calculate tip and total
tip_amount = bill * (tip_percent / 100)
total_amount = bill + tip_amount

# Display results (rounded to 2 decimal places)
print(f"\nTip amount: ${tip_amount:.2f}")
print(f"Total amount to pay: ${total_amount:.2f}")
```

```python
print("🚗 Road Trip Cost Estimator")

# Get user inputs
distance = float(input("Enter total trip distance (km): "))
kpl = float(input("Enter your car's average kilometers per liter (KPL): "))
fuel_price = float(input("Enter fuel price per liter: $"))

# Calculate fuel needed
liters_needed = distance / kpl

# Calculate total cost
total_cost = liters_needed * fuel_price

# Display results
print("\n----- Trip Cost Summary -----")
print(f"Total distance: {distance} km")
print(f"Fuel efficiency: {kpl} km/l")
print(f"Fuel needed: {round(liters_needed, 2)} liters")
print(f"Estimated fuel cost: ${round(total_cost, 2)}")
```
